```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm 

df = pd.read_csv(r'../Chapter2/input/longley.csv', index_col=0)
```

```python
y = df.GNP
X = df.Employed
X = sm.add_constant(X)
```

```python
lr_model = sm.OLS(y,X).fit()
```

```python
# visualizing the predicted line vs the data
# Pick 100 points equally spaced from the min to the max 
n_samples = 100
X_prime = np.linspace(X.Employed.min(), X.Employed.max(), n_samples)
```

```python
X_prime = sm.add_constant(X_prime)
```

```python
y_predicted = lr_model.predict(X_prime)
```

```python
plt.scatter(df.Employed, y)
plt.xlabel("Total Employment")
plt.ylabel("Gross National Product")
plt.plot(X_prime[:,1], y_predicted, 'red', alpha=0.9)
```
